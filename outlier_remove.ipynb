```python
import matplotlib.pyplot as plt

# Assuming db is your DataFrame
plt.figure(figsize=(10, 6))

# Scatter plot with different colors based on the Original_Price column
plt.bar(db['Rent'], db['City'], c=db['Rent'])

# Adding colorbar for reference
plt.colorbar(label='Original Price')

# Adding labels and title
plt.xlabel('Rent')
plt.ylabel('City')
plt.title('Scatter Plot of Rent vs City with Original Price')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a bar chart
plt.figure(figsize=(10, 6))
db['Rent'].value_counts().sort_index().plot(kind='bar', color=['blue', 'green'])
plt.title('Distribution of Cities')
plt.xlabel('Cites')
plt.ylabel('Count')
plt.show()
```
